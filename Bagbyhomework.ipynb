```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
city_data = pd.read_csv("data/city_data.csv")


# Check city data
city_data.head()

```

```python
# File to Load (Remember to change these)

ride_data_to_load = "data/ride_data.csv"
ride_data = pd.read_csv("data/ride_data.csv")
ride_data.head()
```

```python
# Combine the data into a single datasetr_data = city_data.merge(ride_data, on = 'city', how = 'outer')
cdata = city_data.merge(ride_data, on = 'city', how = 'outer')

```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city = cdata.loc[(cdata["type"] == "Urban")]
suburban_city = cdata.loc[(cdata["type"] == "Suburban")]
rural_city = cdata.loc[(cdata["type"] == "Rural")]

# Build the scatter plots for each city types

# Incorporate the other graph properties

# Create a legend

# Incorporate a text label regarding circle size

# Save Figure
```

```python
avg_fare_ucity = urban_city.groupby(['city'])['fare'].mean()
total_rides_ucity = urban_city.groupby(['city']).count()['ride_id']
total_drivers_ucity = urban_city.groupby(['city'])['driver_count'].value_counts()
```

```python

avg_fare_scity = suburban_city.groupby(['city'])['fare'].mean()
total_rides_scity = suburban_city.groupby(['city']).count()['ride_id']
total_drivers_scity = suburban_city.groupby(['city'])['driver_count'].value_counts()
```

```python
avg_fare_rcity = rural_city.groupby(['city'])['fare'].mean()
total_rides_rcity = rural_city.groupby(['city']).count()['ride_id']
total_drivers_rcity = rural_city.groupby(['city'])['driver_count'].value_counts()
```

```python
plt.scatter(total_rides_ucity, avg_fare_ucity,s=total_drivers_ucity*5,
            marker ='o', facecolors ="gold", edgecolors='black',alpha = 0.5, label="Urban")
plt.scatter(total_rides_scity, avg_fare_scity,s=total_drivers_scity*10,
            marker ='o', facecolors ="lightskyblue", edgecolors='black',alpha = 0.5, label="Suburban")
plt.scatter(total_rides_rcity, avg_fare_rcity,s=total_drivers_rcity*20,
            marker ='o', facecolors ="lightcoral", edgecolors='black',alpha =.5, label="Rural")
plt.title("Ride Sharing Data")
plt.xlabel("Total Numbers of Rides Per City")
plt.ylabel("Average Fare($)")
plt.legend(loc='upper right')
plt.savefig("rs.png")
plt.show()
```

## Total Fares by City Type

```python


total_fare = cdata.groupby(['type'])['fare'].sum()
labels = ["Rural","Suburban","Urban" ]
colors = ["lightcoral","lightskyblue","gold"]
explode = (0, 0, 0.1)
plt.title("% of Total Fares By City Types")
plt.pie(total_fare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis("equal")
plt.savefig("% of Total Fares By City Types.png")
plt.show()
```

## Total Rides by City Type

```python
total_rides = cdata.groupby(['type'])['ride_id'].count()
labels = ["Rural","Suburban","Urban" ]
colors = ["lightcoral","lightskyblue","gold"]
explode = (0, 0, 0.1)
plt.title("% of Total Rides By City Types")
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.savefig("% of Total Rides By City Types.png")
plt.show()
```

## Total Drivers by City Type

```python
total_drivers = city_data.groupby(['type'])['driver_count'].sum()
labels = ["Rural","Suburban","Urban" ]
colors = ["lightcoral","lightskyblue","gold"]
explode = (0, 0, 0.1)
plt.title("% of Total Drivers By City Types")
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.savefig("% of Total Drivers By City Types.png")
plt.show()
```
